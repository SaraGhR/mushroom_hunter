```python
train_data = pd.read_csv('train.csv')
train_data
```

```python
train_data.isnull().sum()
```

```python
test_data = pd.read_csv('test.csv')
test_data
```

## preprocessing

```python
spacing = {'c': 0,
           'w' : 1,
           'd' : 2}

train_data['gill-spacing'] = train_data['gill-spacing'].map(spacing)
test_data['gill-spacing'] = test_data['gill-spacing'].map(spacing)
```

```python
size = {'n': 0,
        'b' : 1 }

train_data['gill-size'] = train_data['gill-size'].map(size)
test_data['gill-size'] = test_data['gill-size'].map(size)
```

```python
ring = {'n': 0,
           'o' : 1,
           't' : 2}

train_data['ring-number'] = train_data['ring-number'].map(ring)
test_data['ring-number'] = test_data['ring-number'].map(ring)
```

```python
train_data.columns
```

```python
population = {'a': 0,
           'c' : 1,
           'n' : 2,
           's' : 3,
           'v' : 4,
           'y' : 5}

train_data['population'] = train_data['population'].map(population)
test_data['population'] = test_data['population'].map(population)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for i in [ 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor','gill-attachment', 'gill-color',
          'stalk-shape', 'stalk-root', 'stalk-surface-above-ring','stalk-surface-below-ring', 
          'stalk-color-above-ring','stalk-color-below-ring', 'veil-type', 'veil-color','ring-type',
          'spore-print-color', 'habitat']:
  train_data[i] = label_encoder.fit_transform(train_data[i])
  test_data[i] = label_encoder.fit_transform(test_data[i])

```

```python
train_data.isnull().sum()
```

## scikit-learn
### RandomForestClassifier

```python
y=  train_data[['class']]
```

```python
x = train_data.copy()
x.drop(columns= ['class'], inplace= True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=9,max_depth=9,bootstrap=True,oob_score=True,max_samples=400  )
model.fit(X_train,y_train)
y_pred = model.predict(X_val)
```

```python
# evaluate your model
from sklearn.metrics import recall_score

recall_score(y_val, y_pred,pos_label='p')
```

```python
# evaluate your model
from sklearn.metrics import recall_score
y_pred = model.predict(X_test)
recall_score(y_test, y_pred,pos_label='p')
```
